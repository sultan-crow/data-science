### 10 pt: Which sport has the highest median athlete age? 
To determine this, you might want to plot the median age for each sport. You may want to sort the axes appropriately for your analysis.

```python
# put your code here
filtered_df=df.dropna(subset=['Age'])
#df.shape,filtered_df.shape
sorted_age=filtered_df.sort_values(['Age'], ascending=[False])
sorted_age.groupby(['Games'])['Age'].median()
```

### 10 pt: In which events has the Netherlands won the most medals?

```python
# put your code here
neth_df=df[df['Team']=='Netherlands'].dropna(subset=['Medal'])

#neth_df
sorted_data=neth_df.groupby(['Sport'])[['Medal']].count().reset_index().sort_values(by=['Medal'], ascending=False)
print(sorted_data.iloc[0,:])
```

### 10 pt: Which country won the most medal points in each of the 2016 Summer Olympics?
Rank each country by a score, where 3 points are assigned for a gold medal, 2 points for silver, and 1 point for bronze.

```python
# put your code here
medals=df.dropna(subset=['Medal'])[['Team','Medal']];
medals.loc[medals['Medal']=='Gold','Point']=3
medals.loc[medals['Medal']=='Silver','Point']=2
medals.loc[medals['Medal']=='Bronze','Point']=1
medals.groupby(['Team'])['Point'].agg('sum').reset_index().sort_values(['Point'], ascending=False).reset_index()
#df.apply(lambda x: 3 if x['Medal']=='Gold' else 2, axis=1)
```
